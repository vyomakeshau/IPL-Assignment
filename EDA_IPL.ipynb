# **8. Does winning the toss imply winning the game?**

```python
# Create a new column 'toss_win_match_win' to track if toss winners won the match
df2['toss_win_match_win'] = df2['toss_winner'] == df2['winner']

# Group by toss_win_match_win and count the occurrences of wins and losses
toss_win_outcome = df2['toss_win_match_win'].value_counts()

# Plotting the outcome to visualize
labels = ['Toss Winner Won Match', 'Toss Winner Lost Match']
plt.figure(figsize=(7, 7))
plt.pie(toss_win_outcome, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('Outcome of Matches Based on Toss Winner')
plt.show()

# Display the counts as well
toss_win_outcome
```

# **13. Which team has the highest winning percentage?**

```python
# Step 1: Calculate total matches played by each team
team1_matches = df2['team1'].value_counts()
team2_matches = df2['team2'].value_counts()

# Sum the counts for both team1 and team2 to get total matches played by each team
total_matches_played = team1_matches.add(team2_matches, fill_value=0)

# Step 2: Calculate total matches won by each team
team_wins = df2['winner'].value_counts()

# Step 3: Calculate winning percentage for each team
winning_percentage = (team_wins / total_matches_played) * 100

# Step 4: Find the team with the highest winning percentage
highest_winning_percentage_team = winning_percentage.idxmax()
highest_winning_percentage = winning_percentage.max()

# Display the result
highest_winning_percentage_team, highest_winning_percentage
```

### **Ans - Chennai Super Kings has the highest winning percentage**

# **24. Who has the highest strike rate?**

# **34. Which team has the best scoring run-rate in the first 6 overs?**

```python
# Step 1: Filter deliveries within the first 6 overs
first_six_overs = df1[df1['over'] <= 6]

# Step 2: Group by batting team and calculate total runs and balls faced in the first 6 overs
team_runs_balls = first_six_overs.groupby('batting_team').agg({'total_runs': 'sum', 'ball': 'count'})


# Step 3: Calculate the run rate for each team
team_runs_balls['run_rate'] = (team_runs_balls['total_runs'] / team_runs_balls['ball']) * 6

# Step 4: Find the team with the highest run rate
best_run_rate_team = team_runs_balls['run_rate'].idxmax()
highest_run_rate = team_runs_balls['run_rate'].max()

# Display the result
print(f"The team with the best scoring run rate in the first 6 overs is {best_run_rate_team} with a run rate of {highest_run_rate:.2f}.")
team_runs_balls

```
